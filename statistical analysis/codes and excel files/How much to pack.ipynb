# How much to pack

```r
library(repr) # to set the graph sizes
```

```r
options(repr.plot.width=7, repr.plot.height=5)
mean = 1005
sd = 5/3
x <- seq(1000,1010,by=0.05)
y <- dnorm(x,mean=mean,sd=sd)
plot(x,y,type='l')
abline(a=NULL, b=NULL, h=NULL, v=mean, col='red')
```

```r
z = qnorm(p=0.01, mean=0, sd=1)
z
```

```r
options(repr.plot.width=7, repr.plot.height=5)
mean = 1003.88
sd = 5/3
x <- seq(995,1010,by=0.05)
y <- dnorm(x,mean=mean,sd=sd)
plot(x,y,type='l')
lower_limit = 995
upper_limit = 1000
polygon(x=c(lower_limit,x[lower_limit<=x & x<=upper_limit],upper_limit), 
        y=c(0,y[lower_limit<=x & x<=upper_limit],0), col='red')
abline(a=NULL, b=NULL, h=NULL, v=mean, col='red')
```

```r
probability = pnorm(q=1000,mean=mean, sd=sd)
probability*100
```
